```python
# problem 1
import numpy as np

xnd_array= np.arange(-50,50.1,0.1)
ynd_array = ((xnd_array*1)/2)+1
print(xnd_array.shape)
print(ynd_array.shape)
```

```python
# problem 2
xynd_array = np.concatenate([xnd_array,ynd_array])
xynd_array = xynd_array.reshape(1001,2)
print(xynd_array.shape)
```

```python
# problem 3
x_change = xnd_array[1:] - xnd_array[:-1]
y_change = ynd_array[1:] - ynd_array[:-1]
gradient = y_change/x_change
print(gradient)
```

```python
# problem 4
import matplotlib.pyplot as plt

ynd_array = ynd_array.reshape(1001,1)

plt.figure(1)
plt.title("linear function")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(ynd_array[0:],ynd_array[:,0])
plt.show()

plt.figure(2)
plt.title("x and gradient")
plt.xlabel('x')
plt.ylabel('gradient')
plt.plot(xnd_array[1:],gradient)
plt.show()
```

```python
# problem 5
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    xnd_array_2 = np.arange(x_range[0],x_range[1],x_range[2])
    ynd_array_2 = function(xnd_array_2)
    xynd_array_2 = np.concatenate([xnd_array_2,ynd_array_2])
    gradient = (ynd_array_2[1:] - ynd_array_2[:-1])/(xnd_array_2[1:] - xnd_array_2[:-1])
    return xynd_array_2, xnd_array_2, ynd_array_2, gradient

def function1(array_x):
    array_y = array_x**2
    return array_y

def function2(array_x):
    array_y = (2*(array_x**2))*(2**array_x)
    return array_y

def function3(array_x):
    array_y = np.sin(array_x**(1/2))
    return array_y

def create_graph(array_x_2, array_y_2, gradient):
    
    array_y_2 = ynd_array.reshape(1001,1)

    plt.figure(1)
    plt.title("linear function")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(array_y_2[0:],array_y_2[:,0])
    plt.show()

    plt.figure(2)
    plt.title("x and gradient")
    plt.xlabel('x')
    plt.ylabel('gradient')
    plt.plot(array_x_2[1:],gradient)
    plt.show()
```

### with function 1

```python
array_xy, array_x_2, array_y_2, gradient_1 = compute_gradient(function1)
create_graph(array_x_2, array_y_2, gradient_1)
```

### with function 2

```python
array_xy_2, array_x_3, array_y_3, gradient_2 = compute_gradient(function2)
create_graph(array_x_3, array_y_3, gradient_2)
```

### with function 3

```python
array_xy_2, array_x_4, array_y_4, gradient_3 = compute_gradient(function3)
create_graph(array_x_4, array_y_4, gradient_3)
```
