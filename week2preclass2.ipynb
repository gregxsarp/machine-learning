```python
# problem 1
import numpy as np

# Fixing random state for reproducibility
np.random.seed(0)

MEAN = [-3,0]
COV = [[1.0],[0.8],[0.8],[1.0]]
nd_cov = np.array(COV).reshape(2,2)
nd_array_of_random_numbers=np.random.multivariate_normal(MEAN, nd_cov, 500).reshape(500,2)
```

```python
# problem 2

import matplotlib.pyplot as plt

plt.scatter(x=nd_array_of_random_numbers[:,0],y=nd_array_of_random_numbers[:,1],s=20, c="blue")

plt.show()
```

```python
# problem 3
import matplotlib.pyplot as plt

# with another approach
first_dimension = []
second_dimension = []

for i in range(len(nd_array_of_random_numbers)):
    first_dimension.append(nd_array_of_random_numbers[i][0])
    second_dimension.append(nd_array_of_random_numbers[i][1])

# first dimension
plt.hist(first_dimension, 50, facecolor='blue')

plt.title("first dimension histogram")
plt.xlim(0, -10)
plt.ylim(0, 50)
plt.grid(True)
plt.show()

# second dimension
plt.hist(second_dimension, 50, facecolor='blue')

plt.title("second dimension histogram")
plt.xlim(5, -5)
plt.ylim(0, 50)
plt.grid(True)
plt.show()
```

```python
# problem 4
import numpy as np
import matplotlib.pyplot as plt

MEAN_2 = (0,-3)
COV_2 = [[1.0],[0.8],[0.8],[1.0]]
nd_cov_2 = np.array(COV_2).reshape(2,2)
nd_array_of_random_numbers_2=np.random.multivariate_normal(MEAN_2, nd_cov_2, 500).reshape(500,2)

plt.scatter(x=nd_array_of_random_numbers[:,0],y=nd_array_of_random_numbers[:,1],s=20, c="blue", label="0")
plt.scatter(x=nd_array_of_random_numbers_2[:,0],y=nd_array_of_random_numbers_2[:,1],s=20, c="orange", label="1")

plt.legend()
plt.show()
```

```python
# problem 5
new_nd_array = np.concatenate((nd_array_of_random_numbers,nd_array_of_random_numbers_2),axis=0)
```
